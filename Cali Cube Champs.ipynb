```python
index = 0

df_list = []

for row in main_event.iterrows():
    if rq.get(main_event['URL'][index]).ok == True:
        df = pd.json_normalize(rq.get(main_event['URL'][index]).json().get('cards').get('mainboard'), sep='_')
        df['ID'] = main_event['ID'][index]
        df['CUBE NAME'] = main_event['CUBE NAME'][index]
        df['Cube Type'] = 'Cali Cube Champ'
        df_list.append(df)
        #print(True)
        index+=1
    else:
        print('Error Encountered At Index:', index)
        index+=1
```

```python
main_event_cubes = pd.concat(df_list)
main_event_cubes.head()
```

```python
def tweak_cubes(main_event_cubes):
    return (main_event_cubes
     .assign(details_cmc=main_event_cubes.details_cmc.fillna(0).astype('int8'),
             is_creature=main_event_cubes.details_type.str.contains('Creature'),
             is_land = main_event_cubes.details_type.str.contains('Land'),
             is_pwer = main_event_cubes.details_type.str.contains('Planeswalker'),
             added_to_cube_on = pd.to_datetime(main_event_cubes.addedTmsp, errors='coerce', unit='ms'),
             composite_id = main_event_cubes['CUBE NAME'] + '-' + main_event_cubes['cardID']
        )
     .astype({'Cube Type': 'category', 'details_cmc': 'int8', 'details_rarity': 'category'})
)

cleaned_data = tweak_cubes(main_event_cubes)
```

```python
cleaned_data.loc[cleaned_data['added_to_cube_on'].isnull(),'added_to_cube_on'] = pd.to_datetime(cleaned_data.loc[cleaned_data['added_to_cube_on'].isnull(),:]['addedTmsp']).dt.tz_localize('UTC').dt.tz_convert('UTC')
cleaned_data['added_to_cube_on'] = pd.to_datetime(cleaned_data['added_to_cube_on'], utc=True)
cleaned_data['days_until_added_to_cube'] = ((cleaned_data['added_to_cube_on'])-pd.to_datetime(cleaned_data['details_released_at']).dt.tz_localize('UTC')).dt.days
```

```python
extract = cleaned_data.loc[:,[
      'index'
    , 'CUBE NAME'
    , 'Cube Type'
    , 'cardID'
    , 'addedTmsp'
    , 'added_to_cube_on'
    , 'details_released_at'
    , 'details_name'
    , 'details_full_name'
    , 'details_artist'
    , 'details_rarity'
    , 'details_color_identity'
    , 'details_colors'
    , 'details_set'
    , 'details_cmc'
    , 'details_parsed_cost'
    , 'details_type'
    , 'details_elo'
    , 'details_popularity'
    , 'details_cubeCount'
    , 'details_loyalty'
    , 'details_power'
    , 'details_toughness'
    , 'is_creature'
    , 'is_land'
    , 'is_pwer'
    , 'days_until_added_to_cube'
]]

extract.head()
```
